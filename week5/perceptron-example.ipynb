# Perceptron Classifier Example
### Exercise: train a classifier for the two semi-circles drawn in the image. This is supervised learning, so split the data points as follows:

1. 70% points for training
2. 30% points for test

```python
from random import random
import matplotlib.pyplot as plt
import math

points = 200

x1 = [0 for _ in range(points)]
y1 = [0 for _ in range(points)]
c1 = [0 for _ in range(points)]

x2 = [0 for _ in range(points)]
y2 = [0 for _ in range(points)]
c2 = [0 for _ in range(points)]

for i in range(points):
    d = -1
    r = 4
    w = 6
    # 0 ~ 180: REDS
    a = random()*math.pi
    x1[i] = math.sqrt(random()) * math.cos(a)*(w/2) + ((-(r+w/2) if(random() < 0.5) else (r+w/2)) * math.cos(a))
    y1[i] = math.sqrt(random()) * math.sin(a)*(w) + (r * math.sin(a)) - d
    c1[i] = 1
    # 180 ~ 360: BLUES
    a = random()*math.pi + math.pi
    x2[i] = (r+w/2) + math.sqrt(random()) * math.cos(a)*(w/2) + ((-(r+w/2)) if(random() < 0.5) else (r+w/2)) * math.cos(a)
    y2[i] = -(math.sqrt(random()) * math.sin(a)*(-w) + (-r * math.sin(a))) + d
    c2[i] = -1

plt.scatter(x1, y1, color="r")
plt.scatter(x2, y2, color="b")
plt.show()
```

```python
X = x1+x2
Y = y1+y2
C = c1 + c2

data = []
for i in range(len(X)):
  data.append((X[i], Y[i]))
data
```

```python
x_train = data[:int(len(data)*0.7)] # features
y_train = C[:int(len(data)*0.7)] # class
x_test = data[len(data) - int(len(data)*0.3):] 
y_test = C[len(data) - int(len(data)*0.3):]
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train, y_train)
```

```python
y_pred = list(model.predict(x_test))
y_pred
```

```python
error = 0
for i in range(len(y_pred)):
     if y_pred[i] != y_test[i]:
            error += 1
error = (error / len(y_pred))*100
print(f"Accuracy: {100 - error}%")
```
